### Regression tutorial is implemented by transfer learning 'mobilenet-v3-small' CNN based model to detect face landmarks

```python
from torch.utils.data import Dataset, DataLoader, random_split
from torch.nn import Linear, Module, Sequential, ReLU, MSELoss, Dropout, Flatten, AdaptiveAvgPool2d
from torch.optim import Adam
from torchvision.models import mobilenet_v3_small
import torch
from torchvision import transforms
```

```python
cnn_model = mobilenet_v3_small(weights="IMAGENET1K_V1")
```

Freezing layers to preserve pretrained parameters and unfreezing last layers to adapt more flexibly to this task

```python
for block in cnn_model.features.children():
    for params in block.parameters():
        params.requires_grad = False

last_block = list(cnn_model.features.children())[-2:]
for block in last_block:
  for each in block.parameters():
      each.requires_grad = True
```

Model

```python
class FaceLandmark(Module):
    def __init__(self,cnn_model):
       super().__init__()
       self.cnn_model = cnn_model
       self.cnn_model.classifier = Sequential(
           Linear(in_features=576, out_features=256, bias=True),
           ReLU(),
           Dropout(0.2),
           Linear(in_features=256, out_features=136, bias=True)
       )
    def forward(self, x):
       x = self.cnn_model(x)
       return x
```

```python
# model
face_landmark_model = FaceLandmark(cnn_model)

# loss function
loss_func = MSELoss()

# optimizer
optimizer = Adam(params=face_landmark_model.parameters(), lr=0.001)
```

* convert images files to PIL
* rescale x and y

x' = x * (new_width  / original_width) <br>
y' = y * (new_height / original_height)
* create custom dataset and load it using dataloader

```python
from PIL import Image
import numpy as np
import os

collection = []
clip = 224

def rescale(ax, n_w, o_w):
    return float(ax) * (float(n_w) / float(o_w))

for each in root_xml.findall("images/image"):
   file = each.get("file")
   img = Image.open(os.path.join(root, file))
   rows = {
       "image" : os.path.join(root, file),
       "points" : np.array([])
   }
   for box in each.findall("box"):
       width, height = img.size
       for part in box.findall("part"):
           scaled_x = rescale(part.get("x"), clip, width)
           scaled_y = rescale(part.get("y"), clip, height)
           rows["points"] = np.append(rows["points"], scaled_x)
           rows["points"] = np.append(rows["points"], scaled_y)

   collection.append(rows)


```

```python
import matplotlib.pyplot as plt

image = Image.open(collection[0]["image"])
# image = collection[0]["image"]
img = image.resize((224,224), Image.Resampling.LANCZOS)

points = collection[0]["points"]
x_points = points[0::2]
y_points = points[1::2]

plt.imshow(img)
plt.scatter(x_points, y_points)
```

```python
transformations = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.Lambda(lambda img: img.convert("RGB")),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                        std=[0.229, 0.224, 0.225])

])
```

```python
class face_points(Dataset):
     def __init__(self, transform, collection):
         self.transform = transform
         self.collection = collection

     def __len__(self):
        return len(self.collection)

     def __getitem__(self, index):

        PIL = Image.open(collection[index]["image"])
        points = collection[index]["points"]

        img_tensor = self.transform(PIL) # apply transformation
        points_tensor = torch.from_numpy(points) / 244 # Normalizing points

        return img_tensor, points_tensor

face_detection_dataset = face_points(transformations, collection)
```

```python
train_dataset_len = int(len(face_detection_dataset)*0.8)
test_dataset_len = len(face_detection_dataset) - train_dataset_len
train_dataset, test_dataset = random_split(face_detection_dataset, lengths=[train_dataset_len, test_dataset_len])
```

```python
train_dataset = DataLoader(train_dataset, batch_size=32, shuffle = True)
test_dataset = DataLoader(test_dataset, batch_size=32, shuffle=True)
```

```python
next(iter(train_dataset))[0].shape
```

Training & Validation

```python
def train_loop(model, train_dataset, loss_func, optimizer):
    no_sample = len(train_dataset.dataset)
    model.train()
    for batch, (x,y) in enumerate(train_dataset):
        x, y = x.float(), y.float()
        prediction = model(x)
        loss = loss_func(prediction, y)
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()
        if batch%30 == 0:
           print(f"training loss : {loss.item():.5} | completed : [{x.shape[0]*batch}/{no_sample}]")

def validation_loop(model, test_dataset, loss_func):
    no_batch = len(test_dataset)
    model.eval()
    loss = 0
    with torch.no_grad():
         for x,y in test_dataset:
             x, y = x.float(), y.float()
             prediction = model(x)
             loss += loss_func(prediction, y).item()
    loss /= no_batch
    print(f"validation loss : {loss:.5}")
    return loss
```

Early Stopping -> prevents overfitting

```python
class EarlyStopping:
     def __init__(self, patience = 2, min_delta = 0.001):
         self.count = 0
         self.patience = patience
         self.min_delta = min_delta
         self.minimum = float("inf")

     def stop(self, validation):
          if validation < self.minimum - self.min_delta:
              self.minimum = validation
              self.count = 0
          else:
              self.count+=1
              if self.count >= self.patience:
                 return True
          return False
```

Training & Validation Cycle

```python
epoch = 8
early_stop = EarlyStopping()
for i in range(1, epoch+1):
    print(f"Epoch : {i}")
    train_loop(face_landmark_model, train_dataset, loss_func, optimizer)
    validation = validation_loop(face_landmark_model, test_dataset, loss_func)
    if early_stop.stop(validation):
       break
```

FAST R-CNNstyle models employ two heads: a classification head optimized with binary cross-entropy (BCE) with logits loss and a regression head optimized with mean squared error (MSE) loss. The model first detects multiple faces, selects the top N proposals, crops the corresponding regions, and performs regression on each proposal
